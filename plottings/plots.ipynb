```python
import numpy as np 
import matplotlib.pyplot as plt 
   
small = [30.839546165, 28.747145839, 24.600546672] 
big = [3657.000821699, 4426.857946943, 1578.107532743] 
  
n=3
r = np.arange(n) 
width = 0.33
  
  
plt.bar(r, small
, color = 'b', 
        width = width, edgecolor = 'black', 
        label='2048x2048') 
plt.bar(r + width, big, color = 'g', 
        width = width, edgecolor = 'black', 
        label='8192x8192') 
  
plt.xlabel("Loop Orders") 
plt.ylabel("Time taken") 
plt.title("Matrix Multiplication (4KB pages)") 
  
# plt.grid(linestyle='--') 
plt.xticks(r + width/2,['i,j,k','j,i.k','k,i,j']) 
plt.legend() 
  
plt.show() 
```

```python
import matplotlib.pyplot as plt

# Data for each trace
traces = {
    "Trace 1": {
        "IPC": [2.05946, 2.05939, 2.05927],
        "Accuracy": [99.9976, 99.9973, 99.997]
    },
    "Trace 2": {
        "IPC": [0.406576, 0.406312, 0.404121],
        "Accuracy": [99.3107, 99.3036, 99.1609]
    },
    "Trace 3": {
        "IPC": [0.362922, 0.362615, 0.361268],
        "Accuracy": [97.9377, 97.8587, 97.5394]
    }
}

# Plotting
for trace, data in traces.items():
    fig, ax1 = plt.subplots()

    ax1.set_xlabel('Sample')
    ax1.set_ylabel('IPC', color='tab:blue')
    ax1.plot(range(1, 4), data["IPC"], 'o-', color='tab:blue', label='IPC')
    ax1.tick_params(axis='y', labelcolor='tab:blue')

    ax2 = ax1.twinx()
    ax2.set_ylabel('Accuracy (%)', color='tab:orange')
    ax2.plot(range(1, 4), data["Accuracy"], 's-', color='tab:orange', label='Accuracy')
    ax2.tick_params(axis='y', labelcolor='tab:orange')

    fig.tight_layout()
    plt.title(trace)
    plt.show()
```
